```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv("glassdoor_jobs.csv")
print(df.shape)
df.head(10)
```

## Salary cleaning

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'hour' in x.lower() else 0)
df.hourly.head()
```

```python
df['emp_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer' in x.lower() else 0)
df.emp_provided.head()
```

```python
df.drop(df[df['Salary Estimate'] == '-1'].index, inplace=True)
df.shape
```

```python
#53ùêæ‚àí 91K (Glassdoor est.)
df['salary_trimmed'] = df.apply(lambda x: re.sub(r'.*\$(\d+)-\$(\d+).*', r'\1-\2', x['Salary Estimate']) if x['hourly'] 
                                else re.sub(r'.*\$(\d+)K\s*-\s*\$(\d+)K.*', r'\1-\2', x['Salary Estimate']), axis=1)
df.salary_trimmed.head(30)
```

```python
df['min_salary'] = df.apply(lambda x: float(x['salary_trimmed'].split('-')[0])*2 if x['hourly']
                                      else float(x['salary_trimmed'].split('-')[0]), axis=1)
```

```python
df['max_salary'] = df.apply(lambda x: float(x['salary_trimmed'].split('-')[1])*2 if x['hourly']
                                      else float(x['salary_trimmed'].split('-')[1]), axis=1)
```

## State

```python
df['State'] = df['Location'].apply(lambda x: x.split(',')[1])
```

## Age

```python
df['Age'] = 2021 - df['Founded']
df['Age'].head()
```

## Headquarters same as location

```python
df['work_in_headquarter'] = df['Location'] == df['Headquarters']
df['work_in_headquarter'].head()
```

## n_competitors

```python
df['n_competitors'] = df['Competitors'].apply(lambda x: len(x.split(',')) if x != -1
                                                        else -1)
```

## Revenue

```python
df['Revenue'].value_counts()
```

```python
d = {'Unknown / Non-Applicable': 0, 'Less than $1 million (USD)': 1, '$1 to $5 million (USD)': 2,
     '$5 to $10 million (USD)': 3, '$10 to $25 million (USD)': 4, '$25 to $50 million (USD)': 5,
     '$50 to $100 million (USD)': 6, '$100 to $500 million (USD)': 7, '$500 million to $1 billion (USD)': 8,
     '$1 to $2 billion (USD)': 9, '$2 to $5 billion (USD)': 10, '$5 to $10 billion (USD)': 11, '$10+ billion (USD)': 12}
df.drop(df[df['Revenue'] == 1].index, inplace=True)
df['Revenue_class'] = df['Revenue'].map(d)
```

## Size

```python
df['Size'].value_counts()
```

```python
d = {'Unknown': 0, '1 to 50 employees': 1, '51 to 200 employees': 2, '201 to 500 employees': 3,
     '501 to 1000 employees': 4, '1001 to 5000 employees': 5, '5001 to 10000 employees': 6, '10000+ employees': 7}
df.drop(df[df['Size'] == -1].index, inplace=True)
df['Size_class'] = df['Size'].map(d)
```

## Drop unnamed: 0

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
df.to_csv('glassdoor_jobs_refined.csv')
```
